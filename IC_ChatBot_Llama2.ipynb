```python
#Loading Model & Tokenizer
import transformers
from transformers import AutoTokenizer
import torch

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token = True) #it means verify my token & download my repository
```

```python
#Creating Llama pipeline
from transformers import pipeline #pipeline-end-to-end setup for using the LLMs

llama_pipeline = pipeline(
    "text-generation",
    model=model,
    torch_dtype = torch.float16,
    device_map = 'auto',
)
```

```python
# Getting Responses
def get_llama_response(prompt: str) -> None:
  """
  Generate a response form the Llama model.

  Parameters:
      prompt (str): The user's input/question for the model.

      Returns:
        None: Prints the generated response.
  """
  sequences = llama_pipeline(
      prompt,
      do_sample = True,
      top_k = 10,
      num_return_sequences = 1, # Ensure only one response is returned
      eos_token_id = tokenizer.eos_token_id,
      max_length = 256,
  )
  print("Chatbot:", sequences[0]['generated_text'])

prompt = 'I like "Modern family" and "Desparate Wives". Do you have any recommendations of other similar shows I might like?\n'
get_llama_response(prompt)
```

```python
prompt = " I am learning Data Science with Gen AI. I am from non technical background. Give me five tips to be a pro in this field."
get_llama_response(prompt)
```

```python
#Making it Conversational

while True:
  user_input = input("You: ")
  if user_input.lower() in ["bye", "quit", "exit"]:
    print("Chatbot: Goodbye!")
    break
  get_llama_response(user_input)
```
