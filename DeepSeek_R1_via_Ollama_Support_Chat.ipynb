Interactive Chat Script

```python
# Generating the script
import ollama

# Configuration: Changing the model name
MODEL = "deepseek-r1:1.5b"

# System prompt to define Agent's behavior
messages = [
    {"role":"system", "content":"You are a helpful, empathetic support agent for a tech company called 'CloudStream'. Your goal is to resolve issues quickly and keep the customer happy."}
            ]
print(f"---Chat with DeepSeek Support Agent ({MODEL})---")
print("Type 'exit' or 'quit' to stop the conversation.\n")

while True:
  # 1. Get user input (Customer)
  user_input = input("Customer: ")

  if user_input.lower() in ["exit", "quit"]:
    print("Ending Session....")
    break

  # 2. Add customer message to history
  messages.append({"role":"user","content": user_input})

  # 3. Get model response (Agent)
  print("Agent: ", end="", flush=True)

  response_content = ""
  # Streaming the response makes it feel like a real-time chat
  stream = ollama.chat(model = MODEL, messages = messages, stream = True)

  for chunk in stream:
    content = chunk['message']['content']
    response_content += content
    print(content, end="", flush=True)

  # 4. Add agent's response to history for context
  messages.append({"role": "assistant", "content": response_content})
  print("\n" + "-"*30)
```

*The System Prompt defines the AI's 'personality' and constraints.*
*This ensures the model stays in character as a professional agent
rather than a general-purpose assistant.*


Summary and Insights:

```python
import ollama
import re

def extract_insights(chat_history):
    # Combine the chat history into a single string for the model to analyze
    formatted_chat = ""
    for msg in chat_history:
        if msg['role'] != 'system':
            formatted_chat += f"{msg['role'].capitalize()}: {msg['content']}\n"

    # The Analysis Prompt
    analysis_prompt = f"""
    Analyze the following customer support conversation and provide:
    1. Summary: A 2-sentence overview of the interaction.
    2. Sentiment: Was the customer Happy, Frustrated, or Neutral?
    3. Key Issue: What was the specific problem?
    4. Resolution: Was the issue resolved? (Yes/No)
    5. Action Items: What should the company do next?

    CONVERSATION:
    {formatted_chat}
    """

    print("\n--- Generating Insights ---")
    response = ollama.chat(model=MODEL, messages=[{'role': 'user', 'content': analysis_prompt}])

    # Clean up DeepSeek reasoning tags if they appear
    full_text = response['message']['content']
    clean_text = re.sub(r'<think>.*?</think>', '', full_text, flags=re.DOTALL).strip()

    return clean_text

# Usage:
insights = extract_insights(messages)
print(insights)
```
